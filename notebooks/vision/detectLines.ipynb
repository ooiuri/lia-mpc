```python
# pip install opencv-python
import cv2
import numpy as np
 
# Read image
image = cv2.imread('assets/01/frame0000.jpg')
 
# Convert image to grayscale
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
 
# Use canny edge detection
edges = cv2.Canny(gray,50,150,apertureSize=3)
avg_x = []
y_index = 0
edges
for y in edges:
    print('y: ', y)
    height, width = edges.shape[:2]
    avg_x = int(np.average(y, weights=range(len(y))) + width/2)
    print('avg_x: ', avg_x)
    # cv2.circle(edges,(avg_x,y_index),2,(255,255,255), 1)
    y_index = y_index + 1
# # Apply HoughLinesP method to 
# # to directly obtain line end points
lines_list =[]
lines = cv2.HoughLinesP(
            edges, # Input edge image
            10, # Distance resolution in pixels
            np.pi/180, # Angle resolution in radians
            threshold=200, # Min number of votes for valid line
            minLineLength=10, # Min allowed length of line
            maxLineGap=25 # Max allowed gap between line for joining them
            )
# resolution = [640, 480]
# y_list = np.linspace(0, 640, 5)

# # Iterate over points
for points in lines:
      # Extracted points nested in the list
    x1,y1,x2,y2=points[0]
    # Draw the lines joing the points
    # On the original image
    cv2.line(image,(x1,y1),(x2,y2),(0,255,0),2)
    # Maintain a simples lookup list for points
    lines_list.append([(x1,y1),(x2,y2)])
     
# Save the result image
cv2.imwrite('detectedLines2.png',image)
edges
```
